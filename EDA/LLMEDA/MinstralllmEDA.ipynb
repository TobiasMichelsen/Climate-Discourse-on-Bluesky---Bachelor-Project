```python
#Read in file
input_path = "../../hpc_scripts/LLM/Minstral/TEMP_HPC_FILES"
filename = "LLM_conf_simTex.pkl" 

df = pd.read_pickle(f"{input_path}/{filename}")

#peek into one specific cliuster
df.head()
```

```python
def clean_llm_output(raw_text):
    if not isinstance(raw_text, str):
        return {}
    # Remove leading numbers or indexes like '3 ' or '0 '
    raw_text = raw_text.strip()
    match = re.search(r"\{.*\}", raw_text)
    if not match:
        return {}
    json_like = match.group()
    try:
        parsed = json.loads(json_like)
        return parsed
    except json.JSONDecodeError:
        return {}

# Apply cleaning to your DataFrame
df['parsed'] = df['response'].apply(clean_llm_output)


```

```python
df_cl = df.loc[df.cluster == 1]
print(df_cl.shape)
df_cl.head(10)
```

```python
#see topic distribution over whole df

df = df[df.parsed != {}]
predictions = df.parsed
# Assuming `predictions` is a list of dicts: [{"Fossil Fuels": 0.9, "Politics": 0.8}, {...}, ...]
# Extract all keys (topics) from each dict
all_topics = [topic for pred in predictions for topic in pred.keys()]
topic_counts = Counter(all_topics)

# Optional: Convert to percentage
total = len(all_topics)
topic_distribution = {k: v / total * 100 for k, v in topic_counts.items()}
topic_distribution
```

```python
#see co occurence

co_occurrence = defaultdict(int)

for pred in predictions:
    topics = list(pred.keys())  # Get the list of predicted topics for this text
    for pair in combinations(sorted(topics), 2):  # only pairs
        co_occurrence[pair] += 1

# Convert to DataFrame for better visualization
import pandas as pd
co_df = pd.DataFrame([{"Topic 1": k[0], "Topic 2": k[1], "Count": v} for k, v in co_occurrence.items()])
print(co_df)
```
