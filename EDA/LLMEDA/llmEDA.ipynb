```python

df = pd.read_pickle("LLM_classified_df.pkl")
cluster_topic_counts = defaultdict(Counter)

for i in df.cluster.unique():
    df_cl = df[df.cluster == i]
    for j in range(df_cl.shape[0]):
        row = df_cl.iloc[j]
        output = row.response.split("\n")
        first = output[0]
        first_list = first.split(" ")
        try:
            topic = first_list[1]
        except:
            continue
        cluster_topic_counts[int(i)][topic] += 1
```

```python
cluster_topic_counts.items()
```

```python
for cluster, topic_counts in cluster_topic_counts.items():
    print(f"Cluster {cluster}:")
    for topic, count in topic_counts.items():
        print(f"  {topic}: {count}")
```

```python
# Assume cluster_topic_counts is a dict of Counters
df_summary = pd.DataFrame.from_dict(cluster_topic_counts, orient='index').fillna(0).astype(int)
```

```python
df_summary['prediction'] = df_summary.idxmax(axis=1)
df_summary = df_summary[['prediction'] + [col for col in df_summary.columns if col != 'prediction']]
df_summary.head()
```

```python
#One script to rule them all
df = pd.read_pickle("LLM_df_few_shot.pkl")
cluster_topic_counts = defaultdict(Counter)

for i in df.cluster.unique():
    df_cl = df[df.cluster == i]
    for j in range(df_cl.shape[0]):
        row = df_cl.iloc[j]
        output = row.response.split("\n")
        first = output[0]
        first_list = first.split(" ")
        try:
            topic = first_list[1]
        except:
            continue
        cluster_topic_counts[int(i)][topic] += 1
        
df_summary = pd.DataFrame.from_dict(cluster_topic_counts, orient='index').fillna(0).astype(int)
df_summary['prediction'] = df_summary.idxmax(axis=1)
df_summary = df_summary[['prediction'] + [col for col in df_summary.columns if col != 'prediction']]
print(df_summary.head())
df_summary.groupby("prediction").size()
```

```python

def extract_prediction(response):
    # Regular expression to match the category part before the first ' - '
    match = re.match(r'\d+\.\s*(\w+)', response)
    if match:
        print(match.group(1))
        return match.group(1)
    else:
        return None

# Apply the function to create a new column with predictions
df['prediction'] = df['response'].apply(extract_prediction)

# Display the dataframe with the new 'prediction' column
df.groupby("prediction").size()
```
