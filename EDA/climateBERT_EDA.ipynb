```python
def find_avg_text_len(score, dataf):
    text_lengths_df = dataf[dataf["score"] <= score]["text"].apply(len)
    sum_text_len = text_lengths_df.sum()
    avg_text_len = sum_text_len/len(text_lengths_df)
    return avg_text_len
```

```python
#average text length between score values
def find_avg_text_len_between(score, dataf):
    len_collection = 0
    dataf = dataf[(dataf["score"] > score) & (dataf["score"] <= score+0.05)]
    text_lengths_df = dataf["text"].apply(len)
    sum_text_len = text_lengths_df.sum()
    len_collection += int(len(dataf))
    avg_text_len = sum_text_len/len(dataf)
    # print(dataf["text"].apply(len).describe())
    return avg_text_len
```

```python
#maximize number of posts above some score_threshold  while minimizing total number of posts
def limit_analysis(dataf, min_score=0.95):
    total_high_score = len(dataf[dataf["score"] >= min_score])

    print(f"\nTotal posts with score ≥ {min_score}: {total_high_score}\n")
    print("TextLen | TotalPosts | HighScorePosts | Coverage% | Retention%")
    print("--------|------------|----------------|-----------|------------")

    for length_thresh in range(0, 300, 1):
        filtered = dataf[dataf["text"].apply(len) > length_thresh]
        total = len(filtered)
        high_score = len(filtered[filtered["score"] >= min_score])

        if total > 0:
            coverage = round((high_score / total) * 100, 2)
        else:
            coverage = 0.0
        retention = round((high_score / total_high_score) * 100, 2)

        print(f"{length_thresh:7} | {total:10} | {high_score:14} | {coverage:9}% | {retention:10}%")
```

```python
def collect_score_retention_data(dataf, min_score=0.95, step=10, max_len=100):
    results = []

    total_data_len = len(dataf)
    max_score_posts = len(dataf[dataf["score"] >= min_score])

    for length_thresh in range(0, max_len + 1, step):
        filtered_df = dataf[dataf["text"].apply(len) > length_thresh]
        total = len(filtered_df)
        high_score = len(filtered_df[filtered_df["score"] >= min_score])

        # Avoid divide-by-zero
        coverage = round((high_score / max_score_posts) * 100, 2) if max_score_posts > 0 else 0
        retention = round((total / total_data_len) * 100, 2) if total_data_len > 0 else 0

        results.append({
            "text_length_threshold": length_thresh,
            "total_posts": total,
            "high_score_posts": high_score,
            "coverage_%": coverage,
            "retention_%": retention
        })

    return pd.DataFrame(results)
```

```python
def plot_post_counts(df_analysis):
    plt.figure(figsize=(10, 6))
    
    plt.plot(df_analysis["text_length_threshold"], df_analysis["total_posts"], label="Total Posts", marker='o')
    plt.plot(df_analysis["text_length_threshold"], df_analysis["high_score_posts"], label="High Score Posts (score ≥ 0.95)", marker='o')
    
    plt.xlabel("Text Length Threshold")
    plt.ylabel("Number of Posts")
    plt.title("Number of Posts vs Text Length Threshold")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
