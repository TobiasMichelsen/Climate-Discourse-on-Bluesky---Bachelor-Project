```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
with open("../data/processed/Climate_labeled_posts.json") as f:
    df = pd.read_json(f)
df.head()
```

```python
df_yes = df[df["label"] == "yes"]
df_no = df[df["label"] == "no"]
```

```python
counts = df["label"].value_counts()
print(counts)
print(f"\nall rows: {len(df)}\n")
print(f"percentage no: {counts.iloc[0]/len(df)}")
print(f"percentage yes: {counts.iloc[1]/len(df)}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 1 row, 3 columns
axes = axes.flatten()

# Plot 1: All scores
axes[0].hist(df["score"], bins=50, color='skyblue', edgecolor='black')
axes[0].set_title("All labels")
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Frequency")
axes[0].grid(True)

# Plot 2: yes scores
axes[1].hist(df_yes["score"], bins=50, color='green', edgecolor='black')
axes[1].set_title("yes label")
axes[1].set_xlabel("Score")
axes[1].set_ylabel("Frequency")
axes[1].grid(True)

# Plot 3: no scores
axes[2].hist(df_no["score"], bins=50, color='red', edgecolor='black')
axes[2].set_title("no label")
axes[2].set_xlabel("Score")
axes[2].set_ylabel("Frequency")
axes[2].grid(True)
fig.delaxes(axes[3])
plt.tight_layout()
plt.show()
```

```python
print(f"above .9 score posts: {len(df[df["score"] >= 0.9])}")
print(f"percentage of total: {len(df[df["score"] >= 0.9])/len(df)}")

print(f"\nabove .9 score yes posts: {len(df_yes[df_yes["score"] >= 0.9])}")
print(f"percentage of total yes posts: {len(df_yes[df_yes["score"] >= 0.9])/len(df_yes)}")

print(f"\nabove .9 score no posts: {len(df_no[df_no["score"] >= 0.9])}")
print(f"percentage of total no posts: {len(df_no[df_no["score"] >= 0.9])/len(df_no)}")

print(f"\nabove .99 score posts: {len(df[df["score"] >= 0.99])}")
print(f"percentage of total: {len(df[df["score"] >= 0.99])/len(df)}")

print(f"\nabove .99 score yes posts: {len(df_yes[df_yes["score"] >= 0.99])}")
print(f"percentage of total yes posts: {len(df_yes[df_yes["score"] >= 0.99])/len(df_yes)}")

print(f"\nabove .99 score no posts: {len(df_no[df_no["score"] >= 0.99])}")
print(f"percentage of total no posts: {len(df_no[df_no["score"] >= 0.99])/len(df_no)}")
```

```python
#Getting average text length for all data for score above .9 and .99

df_9 = df[df["score"] >= .9]
df_99 = df[df["score"] >= .99]


text_lengths = df["text"].apply(len)
print(text_lengths.sum()/len(df))

text_lengths = df_9["text"].apply(len)
print(text_lengths.sum()/len(df_9))

text_lengths = df_99["text"].apply(len)
print(text_lengths.sum()/len(df_99))
```

```python
#Getting average text length for yes data for score above .9 and .99

df_yes_9 = df_yes[df_yes["score"] >= .9]
df_yes_99 = df_yes[df_yes["score"] >= .99]


text_lengths = df_yes["text"].apply(len)
print(text_lengths.sum()/len(df_yes))

text_lengths = df_yes_9["text"].apply(len)
print(text_lengths.sum()/len(df_yes_9))

text_lengths = df_yes_99["text"].apply(len)
print(text_lengths.sum()/len(df_yes_99))
```

```python
#Getting average text length for no data for score above .9 and .99

df_no_9 = df_no[df_no["score"] >= .9]
df_no_99 = df_no[df_no["score"] >= .99]


text_lengths = df_no["text"].apply(len)
print(text_lengths.sum()/len(df_no))

text_lengths = df_no_9["text"].apply(len)
print(text_lengths.sum()/len(df_no_9))

text_lengths = df_no_99["text"].apply(len)
print(text_lengths.sum()/len(df_no_99))
```
