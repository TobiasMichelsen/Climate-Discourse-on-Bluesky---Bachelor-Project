```python
import sqlite3
import pandas as pd
```

```python
# Connect to the .db file in the 'data' folder
conn = sqlite3.connect('../database/firstFetch.db')
cur = conn.cursor()
# Example: List tables in the database
cur.execute("SELECT * FROM sqlite_sequence")
print(f"general:\n{cur.fetchall()}")

print("Metadata example:\nColumn ID (index), Column name, Data type, Whether it allows NULL values, Default values (if any), Whether it's a primary key\n")
cur.execute("PRAGMA table_info(posts)")
columns = cur.fetchall()
print(f"Posts table metadata: \n {columns}")

conn.close()
```

```python
# POSTS
conn = sqlite3.connect('../database/firstFetch.db')
cur = conn.cursor()
cur.execute("PRAGMA table_info(posts)")
columns = cur.fetchall()
column_names = [col[1] for col in columns]
print(f"Posts column names:\n {column_names}")
conn.close()
```
