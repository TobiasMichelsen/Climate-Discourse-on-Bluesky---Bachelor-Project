```python
import pandas as pd
```

```python
df_run1 = pd.read_json("../run1/run1_result.json",lines=True)
df_clean = pd.read_json("run2_result_cleaned.json",lines=True)
df_0 = df_run1[df_run1.topic==0]
```

```python
print(df_run1.shape)
print(df_clean.shape)
print(df_0.shape)
```

```python
df_0_dedup = df_0.drop_duplicates("cid",keep="first")
df_clean_dedup = df_clean.drop_duplicates("text",keep="first")
```

```python
print(df_0_dedup.shape)
print(df_clean_dedup.shape)

```

```python
df_clean_dedup
df_0_dedup
```

```python
df_clean_dedup = df_clean_dedup[["text", "topic"]]
df_0_dedup = df_0_dedup[["cid","text"]]
```

```python
df_clean_dedup
df_0_dedup
```

```python
df_full = df_clean_dedup.merge(df_0_dedup,on="text",how="left")
```

```python
df_full.to_json("run2_result_clean_cid.json", orient= "records", lines=True)
```
